- 남성 흡연자 조건 필터링 
- tip을 가장 많이 낸 순서로 내림 차순 정렬
- 상위 10개 출력

```python
tips.value_counts('size')
```

```python
tips1 = tips['sex'] == 'Male'
tips2 = tips['smoker'] == 'Yes'
```

```python
#tp1 = tips.loc[tips1 & tips2]
tips.loc[tips1 & tips2].sort_values('tip', ascending = False).head(10)
```

```python
#tp1.sort_values('tip' ,ascending = False).head(10)
```

- 식사인원이 5명 이상인 저녁테이블
- 흡연자가 없는 주말 테이블
- 토탈 bill, tip출력 하는 것임 

```python
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'

cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sat', 'Sun'])

tips.loc[(cond1 & cond2) & (cond3 & cond4)][['total_bill', 'tip']]
# tips.loc[(cond1 & cond2) & (cond3 & cond4), ['total_bill', 'tip']]

#df.loc[condition]['age']
```
