```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')
```

```python
df.head(3
)
```

```python
# 남자 생존률의 평균/ 여자 생존률의 평균 
```

```python
df_c =df.copy()
```

# groupby

```python
df_c.groupby('sex').mean(numeric_only= True)
```

```python
df_c.groupby(['sex', 'pclass']).mean(numeric_only = True)
# 앞에서 부터 내가 보기 싶은 데이터를 적으면 점점 묶여 들어 가는 거임,,,,
```

```python
df_c.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only = True)
```

```python
# 해당은 그룹핑으로 되어 있어 데이터에 접근에 어렵다
```

# reset_index()

```python
df_temp = df_c.groupby(['sex', 'pclass'])[['survived']].mean()
```

```python
df_temp.reset_index()
```

```python
df_c.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean', 'sum'])
```

# pivot_table

```python
df_c.pivot_table(index = 'who', values = 'survived')
```

```python
df_c.pivot_table(index = ['who', 'pclass'], values = 'survived')
```
